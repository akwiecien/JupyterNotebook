```
data = spark.read.load('FileStore/tables/MillionSales.csv', format='com.databricks.spark.csv', header='true')
```

```
display(data.take(5))
```

```
poland = data.select(data["Region"], data["Country"], data["Item Type"].alias("ItemType"), data["Sales Channel"].alias("SalesChannel"), data["Order Priority"].alias("OrderPriority")).filter('Country == "Poland"')
```

```
# we can create table just by saving DataFrame
poland.write.saveAsTable("PolandTable")
#column names cannot have white signs
```

```
spark.sql("DROP TABLE WorldTable")
#spark.sql("CREATE TABLE WorldTable USING CSV OPTIONS (path '/FileStore/tables/MillionSales.csv', header 'true', inferSchema 'true')")
```

```
short = data.select(data["Region"], data["Country"], data["Unit Price"].alias('UnitPrice')).filter("Country == 'Poland'")
```

```
#save DataFrame as Table
short.write.saveAsTable("ShortTable")
```

```
display(short.take(5))
```

```
#insert new record to table
spark.sql("INSERT INTO ShortTable VALUES ('Africa', 'RPA', 100.00)")
```

```
spark.sql("select * from ShortTable where Region not in ('Europe')").show()
```
